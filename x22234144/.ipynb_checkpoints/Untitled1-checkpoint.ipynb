```python
import pandas as pd

df = pd.read_csv('Output.csv')
```

```python
df
```

```python
columns_to_round = ['Open', 'High', 'Low', 'Close', 'Volume']
df[columns_to_round] = df[columns_to_round].round(2)
```

```python
df.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Replace this with the actual name of your DataFrame

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a line plot
plt.figure(figsize=(14, 8))

# Plot 'Total Confirmed cases'
sns.lineplot(x='Date', y='Total Confirmed cases', data=df, label='Total Confirmed cases', color='b')

# Plot 'Death'
sns.lineplot(x='Date', y='Death', data=df, label='Death', color='r')

plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Total Confirmed Cases and Death Over Time')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import MonthLocator

# Assuming your DataFrame is named 'df'
# Replace this with the actual name of your DataFrame

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a line plot
plt.figure(figsize=(14, 8))

# Plot 'Total Confirmed cases'
sns.lineplot(x='Date', y='Total Confirmed cases', data=df, label='Total Confirmed cases', color='b')

# Create a second y-axis (right) for 'High'
ax2 = plt.gca().twinx()
sns.lineplot(x='Date', y='High', data=df, label='High', color='r', ax=ax2)

plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases', color='b')
ax2.set_ylabel('High', color='r')
plt.title('Total Confirmed Cases and High Over Time')
plt.legend(loc='upper left', bbox_to_anchor=(0.7, 0.9))  # Adjust the legend position as needed

# Set the locator for the x-axis ticks to every month
plt.gca().xaxis.set_major_locator(MonthLocator())

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
